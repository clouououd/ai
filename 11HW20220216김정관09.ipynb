```python
# Algorithm Homework(#31)
1. input(문자와 숫자가 섞여있다.)
2. isdigit() 숫자인가요? (반복)
3. 숫자면 int : 이쪽으로
4. 문자면 str : 이쪽으로 
```

Algorithm_Homework(#31).jpg

```python
data = input()
int_data = ''
str_data = ''

for i in data :
  if i.isdigit() :
    int_data += i
  else :
    str_data += i

print(f'문자 : {str_data}\n숫자 : {int_data}')    
```
