```python
잠정적인 최소값의 인덱스를 저장하는 변수 indexMin을 준비한다.
그때그때 마다의 최소값의 첨자를 저장할 indexMin을 만든다.
```

```python
arr[indexMin]와 요소들의 비교처리는 같은 절차의 반복이다. i는 0에서 4까지 순차적으로 변하면서 반복된다.
이 변하는 구조의 점차는 변수 k로 나타내보자
k의 초기값은 1이고 최대 4까지이다.
```

```python
잠정 최소값과 비교하는 요소의 첨자를 변수 k로 치환하여 반복 구조로 만들었다.
이대로라면 k에 무한루프가 발생된다. 반복구조는 언제나 종료 조건이 필수이다.

i가 0일때 k는 1부터 4까지 변하게 된다. k가 5가되는 일은 절대 없다. 따라서
종료 조건은 k < 5 으로 설계한다.
```

```python
최소값과 arr[i]의 데이터를 교환
```

Selection_Sort.jpg

```python
def selection_sort(arr) :
  for i in range(len(arr)-1) :
    indexMin = i
    for k in range(i,len(arr)-1) :
      k = k + 1
      if arr[k] < arr[indexMin] :
        indexMin = k
    w = arr[i]
    arr[i] = arr[indexMin]
    arr[indexMin] = w

  return arr
    
selection_sort([12, 13, 11, 14, 10])
```

```python
# Bubble Sort Algorithms

1. 오른쪽 끝부터 순서대로 인접한 공 2개를 오름차순 정렬한다.
2. 왼쪽 끝 칸부터 순서대로 들어갈 공을 확정시켜 나간다.
# 왼쪽부터 정렬을 시작해도 된다. 큰수가 왼쪽으로 오게하면 내림차순

오른쪽 끝 인접한 공 2개
오른쪽 요소 : i      -> arr[i]
왼쪽 요소   : i - 1  -> arr[-i]

정렬이 이미 되어있는 상태인지 여부를 확인하려면 
'왼쪽 요소는 오른쪽 요소보다 작은가?' : arr[i - 1] < arr[i]
```

```python
(개선사항)
마지막에 i - 1이 0이되고 i가 1이 되면
이미 확정된 arr[0]과 arr[i]을 비교하게 된다. 이것은 쓸데 없는 낭비이다.

(->)
'i > 0' 조건식 부분이 개선이 필요해 보인다.
arr[i]를 확정하는 반복처리부터는 'i > 0'을 'i > 1'로 바꾸어야한다.
반복할 때마다 i > 0, i > 1, i > 2 ... 와 같이 늘려 가도록 변화하는 처리를 추가하자
변수 k로 표현하자. k의 초기값은 0부터 시작한다. i > k
```

Bubble_Sort.jpg

```python
def bubble_sort(arr) :
  for k in range(len(arr)-1) :
    for i in range(len(arr)-1,k,-1) :
      if arr[i-1] > arr[i] :
        w = arr[i-1]
        arr[i-1] = arr[i]
        arr[i] = w
  
  return arr

bubble_sort([5, 3, 4, 1, 2])
```
