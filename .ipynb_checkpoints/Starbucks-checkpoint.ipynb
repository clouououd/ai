```python
# 파이썬으로 브라우저 컨트롤 모듈
from selenium import webdriver

# 크롤링한 데이터 정제 모듈
from bs4 import BeautifulSoup as bs

# 데이터 분석 무듈
import pandas as pd
```

```python
# 파이썬으로 크롬브라우저를 컨트롤해 해당 웹사이트를 열기
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.starbucks.co.kr/store/store_map.do')
```

```python
# 파이썬으로 웹사이트에 버튼 클릭하기
# 버튼에서 마우스 오른쪽 > 검사 , 해당부분에서 마우스 오른쪽 > copy > select
# 부분이 아래의 ''안의 부분
area_btn = '#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a'
driver.find_element_by_css_selector(area_btn).click()
```

```python
seoul_btn = '#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(seoul_btn).click()
```

```python
seoul_all_btn = '#mCSB_2_container > ul > li:nth-child(1) > a'
driver.find_element_by_css_selector(seoul_all_btn).click()
```

```python
# 전체 페이지 읽어오기
html = driver.page_source

# bs으로 데이터 정제
soup = bs(html)
```

```python
# 서울 전체 스타벅스 목록만 읽어오기

starbucks_list = []
```

```python
starbucks_list = soup.select('li.quickResultLstCon')
```

```python
starbucks = starbucks_list[0]
```

```python
# 지점명 (키값 벨류값 이용)
# name = starbucks['data-name']
# 지점명 (코드를 이용)
name = starbucks.select('strong')[0].text
# 위도
lat = starbucks['data-lat']
# 경도
lng = starbucks['data-long']
# 주소
addr = str(starbucks.select('p')[0]).split('<br/>')[0].split('>')[1]
print(type(addr))
# 전화번호
tel = str(starbucks.select('p')[0]).split('<br/>')[1].split('<')[0]

print(name)
print(lat)
print(lng)
print(addr)
print(tel)
```

```python
# 데이터 프레임 생성
starbucks_data = []

for starbucks in starbucks_list :
    # 지점명 (코드를 이용)
    name = starbucks.select('strong')[0].text
    # 위도
    lat = starbucks['data-lat']
    # 경도
    lng = starbucks['data-long']
    # 주소
    addr = str(starbucks.select('p')[0]).split('<br/>')[0].split('>')[1]
    print(type(addr))
    # 전화번호
    tel = str(starbucks.select('p')[0]).split('<br/>')[1].split('<')[0]
    
    starbucks_data.append([name, lat, lng, addr, tel])

df = pd.DataFrame(starbucks_data, columns=['지점명','위도','경도','주소','전화번호'])

df.to_excel('starbucks.xlsx', index=False)
```
