```python
# Algorithm Homework(#27)
# 홀수와 짝수의 개수를 구하는 fnc 함수를 만드세요.
1. 홀수는 2로 나눠서 나머지가 1
2. 짝수는 2로 나눠서 나머지가 1
3. 경우를 만족할 때마다 count+=1
4. 함수 이름은 func
5. 입력은 리스트를 받는다
6. 출력예시) 홀수 {3}개, 짝수 {2}개
```

image.png

```python
def func(li) :
  count_odd = 0
  count_even = 0

  try :
    for i in li :
      if i % 2 == 1 :
        count_odd += 1
      else :
        count_even += 1
  except :
    pass
  print(f'홀수 {count_odd}개, 짝수 {count_even}개')

li = [41, 43, 45, 47, 49]
func(li)
```
