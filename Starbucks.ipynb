```python
# 파이썬으로 브라우저 컨트롤 모듈
from selenium import webdriver

# 크롤링한 데이터 정제 모듈
from bs4 import BeautifulSoup as bs

# 데이터 분석 무듈
import pandas as pd
```

```python
# 파이썬으로 크롬브라우저를 컨트롤해 해당 웹사이트를 열기
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.starbucks.co.kr/store/store_map.do')
```
