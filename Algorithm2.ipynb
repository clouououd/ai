```python
arr = [4, 2, 8, 9, 5]

for i in range(0,5) :
  if arr[i] == 5 :
    print(f'{i}번째 요소가 일치')
    break
  elif i == 4 :
    print('찾지못했습니다.')
```

```python
def linear_search(no, list) :
  for i in range(len(list)) :
    if no == list[i] :
      return i # print(f'{i}번째 요소가 일치')
  return -1 # print('찾지못했습니다.') # 보통 -1 혹은 None으로 처리를 많이 한다

print(linear_search(9, [4, 2 ,3 ,5 ,1]))
```

```python
# Binary Search 이진 탐색법

# 이진 탐색법은 탐색하는 범위를 절반씩 줄여서 탐색하는 방법이다.
# 데이터들이 먼저 오름차순이나 내림차순으로 정렬되어 있어야만 쓸 수 있다.

11, 13, 17, 19, 23, 29, 31
찾는 값은 17

# 방안에 무슨 값이 들어 있는지 알 수 없는 상황
1st 제일 가운데 있는 3번을 열어본다
0   1   2   3   4   5   6  
            -
            19 <- 찾고 있는 17이 아니다
# 순서대로 정렬 되어있어서 아래의 3곳중 하나에 있다.
2nd 
0   1   2
    -
    13 <- 찾고 있는 17이 아니다
3rd
2
-
17 <- 17을 발견

# 이진 탐색법의 알고리즘
1. 가운데 요소를 선택 (평균을 이용한다)
2. 가운데 요소와 데이터와 찾는 데이터를 비교
3. 탐색 범위를 절반으로 줄이는 처리
```

BinarySearch_Search.jpg

```python
arr = [17, 19, 20, 129]

head = 0
tail = len(arr)
center = 0

while arr[center] != 17 :
  if head <= tail :
    center = (head + tail) // 2
    if arr[center] == 17 :
      print(f'{center}번째 요소가 일치')
      break
    elif arr[center] < 17 :
      head = center + 1
    else :
      tail = center - 1
  else :
    print('찾지못했습니다.')
    break
```
