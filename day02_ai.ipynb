```python
###################################################
# 리스트(list)/튜플(tuple)/딕셔너리(dict)/셋(set)
###################################################
# 여러개의 값을 한번에 저장하는 자료형이 필요하다.

# [리스트 list]
# 지금까지 사용한 숫자 하나, 문자 하나 등 개별적, 독립적인 자료가 아닌
# 리스트는 이러한 자료들 여러개를 모아서 사용할 수 있게 해준다.
# 리스트에 넣는 자료를 요소(element)라고 한다.

# 리스트는 []로 구성한다.
list = [1,2,3,4,5]
print(list)

# 리스트는 다양한 자료형의 데이터를 한번에 담을 수 있다.
# *하나의 list에 여러형의 자료를 담는 것은 오류를 일으킬 가능성이 있기 때문에 하나의 형태의 자료만 담는 것이 좋다.
list = [1,2,3,'대한','민국',True,False]
print(list)

# 리스트의 요소를 뽑아내는 법
score = [80, 90, 70, 60]
print(score) # list전체가 출력됨
# 리스트는 인덱싱(indexing)과 슬라이싱(slicing)이 가능하다.
print(score[0]) # 파이썬의 index는 0부터 시작
print(score[0:2]) # 슬라이싱은 0부터 2미만...처럼 끝은 미만을 나타낸다.

# list는 수정/삭제가 가능하다.(mutable) *tuple은 수정/삭제 불가능(immutable)
score[0] = 100
print(score)
score[-1] = 100
print(score)
# score[4] #indexerror 자주 접하게 될 에러(예외)

# 리스트는 연산이 가능하다
list_a = [1, 2, 3]
list_b = [4, 5, 6]
print('list_a =', list_a)
print('list_b =', list_b)
print(list_a + list_b) # 리스트가 합쳐짐
print(list_a*3) # 리스트가 연속되게 합쳐짐
print(len(list_a)) # 리스트 길이(index 개수)

# 리스트에 많이 사용되는 함수들
# append 리스트에 요소를 추가할 때 사용, 끝에만 추가
print(score)
score.append(60) # 제일 끝에 추가됨
print(score)
score.append(70)
print(score)

# insert() 리스트 요소를 추가할 때 사용, 특정 위치를 지정하여 추가
score.insert(0, 10)
print(score)
score.insert(-1, 50)
print(score)

# 리스트에 요소를 여러게 추가할 때
print(list_a)
list_a.extend([4,5,6])
print(list_a)

# 파괴적 처리 vs 비파괴적 처리
# 리스트 연결만으로는 list_a에 어떠한 영향도 주지 않는다.(비파괴적 처리)
# 하지만 extend()를 사용하면 list_a에 직접적으로 영향을 주게된다.(파괴적 처리)
# 기본적으로 자료를 사용할 때에는 비파괴적으로 사용하는 것이 바람직하다.
# 가능하면 원본 데이터를 그대로 유지하고 사용하는 것이 선택지가 더 넓다.
# 파괴적 처리
list_c = [1,2,3]
list_c.extend([4,5,6])
print(list_c) # list_c가 바뀌어짐
#비파괴적 처리
list_c2 = [1,2,3]
list_d = [4,5,6]
print(list_c2+list_d)
print(list_c2) # list_c2가 그대로임

# del[] 리스트에서 인덱스값을 이용하여 데이터 삭제
list_e = [0,1,2,3,4,5]
del list_e[1] # index값을 이용하여 삭제
print(list_e)
del list_e[1:3] # 슬라이싱으로 삭제
print(list_e) 
del list_e[:] # 생략하면 처음부터 끝까지 삭제
print(list_e)

# pop() 리스트에서 마지막 위치 데이터 삭제
list_e2 = [0,1,2,3,4,5]
list_e2.pop()
print(list_e2)
list_e2.pop(1) # 인덱스 값으로 데이터 삭제
print(list_e2)

# remove 값으로 삭제
list_e3 = [1,2,1,2]
list_e3.remove(2) # remove는 요소의 값을 찾아 삭제한다.
                  # 동일한 값이 여러개 있을때는 하나씩 순차적으로 삭제한다.
                  # 리스트에서 중복된 여러개의 값을 제거하려면 반복문을 사용한다.
print(list_e3)

# clear() 리스트 요소를 전부 삭제
list_e4 = [0,1,2,3,4,5]
list_e4.clear()
print(list_e4)

# in / not in 리스트에 존재하는 요소인지를 확인
list_f=[222,434,65,123,547,876]
print(222 in list_f)
print(221 in list_f)
print(222 not in list_f)
print(221 not in list_f)
```

```python
# [딕셔너리 dict (java의 map, JS의 json 과 유사한 자료구조)]
# 리스트가 인덱스를 기반으로 값을 저장한다면 딕셔너리는 키값을 기반으로 값을 저장한다.
# {}를 사용하여 데이터들을 저장한다.
# 삽입/삭제가 가능하다. (mutable)
# 순서가 없으므로 인덱싱과 슬라이싱이 불가능하다.
dict = {
    'name' : '엔드 게임',
    'no' : 1,
    'type' : '히어로 무비',
    'price' : 10000
} # key값 : value값
print(dict)
print(dict['name']) # key값을 이용하여 value값을 불러낸다

# value값의 위치에 list를 넣어서 value값으로 여러 값을 담을 수 있다. 
dict2 = {
    '감독' : ['안소니','루소'],
    '배우' : ['타노스','토르','헐크']
}
print(dict2['감독'])

# 요소변경, 삽입, 수정, 삭제
dict3 = {
    'name' : '건조 망고',
    'type' : '절임 과일',
    'ingredient' : ['망고','설탕','색소'],
    'origin' : '필리핀'
}
print('name :',dict3['name'])
print('type :',dict3['type'])
print('ingredient :',dict3['ingredient'])
print('origin :',dict3['origin'])
# dict의 요소 바꾸기
dict3['name'] = '망고'
print('name :',dict3['name'])
# dict의 요소중 리스트요소 속 값을 불러오기
print(dict3['ingredient'][0])
# 삽입
dict3['price'] = 5000
print(dict3)
# 수정
dict3['price'] = 4000
print(dict3)
# 삭제
del dict3['price']
print(dict3)

# 빈 딕셔너리
dict4 = {}
dict4['name'] = '새이름'
dict4['height'] = 185
dict4['weight'] = 75
print(dict4)
del dict4['weight']
print(dict4)

# dict 자루 등장하는 에러(예외)
# KeyError 리스트에서 없는 인덱스에 접근하려고 하면 IndexError가 발생되는 것처럼 딕에서도 존재하지 않는 키값을 사용하면 예외 또는 에러를 발생시킨다.
dict5 = {
    'name' : '건조 망고',
    'type' : '절임 과일',
    'ingredient' : ['망고','설탕','색소'],
    'origin' : '필리핀'
}
# del dict['nane'] -> 에러

# key값이 dict에 있는지 확인하기
dict6 = {
    'name' : '건조 망고',
    'type' : '절임 과일',
    'ingredient' : ['망고','설탕','색소'],
    'origin' : '필리핀'
}
key = input('>삭제를 원하는 데이터의 키값을 입력하세요 : ')
if key in dict6 :
  print(dict6[key])
else :
  print('존재하지 않는 키에 접근하고 있습니다..')
```
